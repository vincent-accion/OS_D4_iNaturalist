```python
def in_common(df1, df2):
    merged_df = pd.merge(df1, df2, on=['lat_x', 'lat_y'], how='inner')

    return merged_df
```

```python

common_inat_brown_50_temp = in_common(inat_boxes_50, brown_boxes_50)
# display(common_inat_brown_50_temp)

selection_for_2017on = common_inat_brown_50_temp[['lat_x',	'lat_y',	'color_value_x',	'dtb_height_x',	'dtb_width_x',	'RBG_custom_x']].copy()
selection_for_2017on['color_value_x'] = common_inat_brown_50_temp['color_value_y']
selection_for_2017on.columns = ['lat_x', 'lat_y', 'color_value', 'dtb_height', 'dtb_width', 'RBG_custom']
# display(selection_for_2017on)
```

```python
common_inat_brown_50 = common_inat_brown_50_temp[['lat_x',	'lat_y',	'color_value_x',	'dtb_height_x',	'dtb_width_x',	'color_value_y']].copy()
common_inat_brown_50.columns = ['lat_x', 'lat_y', 'color_value_inat', 'dtb_height', 'dtb_width', 'color_value_brown']
common_inat_brown_50['color_value'] = common_inat_brown_50['color_value_brown']/common_inat_brown_50['color_value_brown'].max() + common_inat_brown_50['color_value_inat']/common_inat_brown_50['color_value_inat'].max()
display(common_inat_brown_50)
```

```python
add_boxes_plot(df_start=common_inat_brown_50, title_end='Zones in common between INAT-09-16 and Brown&Al')
```

anyway when calcultaing corr I need to at least normalize by the total number of sightings per dataset

```python
# display(common_inat_brown_50)
common_inat_brown_50['color_value_brown_normalized'] = common_inat_brown_50['color_value_brown']/common_inat_brown_50['color_value_brown'].max()
common_inat_brown_50['color_value_inat_normalized'] = common_inat_brown_50['color_value_inat']/common_inat_brown_50['color_value_inat'].max()

# Calculate the correlation value
correlation_value = common_inat_brown_50['color_value_brown_normalized'].corr(common_inat_brown_50['color_value_inat_normalized'])

print(f'{correlation_value = }')

# Define the custom trendline function
def custom_trendline(x, slope=correlation_value):
    return slope * x

# make custom corr_like function taking the difference between the normalized
difference_list = []
for row in common_inat_brown_50.index:
    difference_list.append(abs(list(common_inat_brown_50['color_value_brown_normalized'])[row] - list(common_inat_brown_50['color_value_inat_normalized'])[row]))
    
common_inat_brown_50['differences_pp'] = difference_list

fig_normalized = px.scatter(common_inat_brown_50,
                            x='color_value_brown_normalized',
                            y='color_value_inat_normalized',
                            trendline_scope="overall",
                            trendline_color_override='blue',
                            color='differences_pp'
                            )
fig_normalized.update_traces(showlegend=False)

fig_normalized.add_scatter(x=common_inat_brown_50['color_value_brown_normalized'],
                           y=custom_trendline(common_inat_brown_50['color_value_brown_normalized'], slope=1),
                           mode='lines',
                           line=dict(color='black', width=2, dash = 'dash'),
                           name='Theoretical Perfect Correlation')

fig_normalized.add_scatter(x=common_inat_brown_50['color_value_brown_normalized'],
                           y=custom_trendline(common_inat_brown_50['color_value_brown_normalized']),
                           mode='lines',
                           line=dict(color='orange', width=2),
                           name='Actual Correlation (R = 0.762) - BEFORE SELECTION')

fig_normalized.add_scatter(x=common_inat_brown_50['color_value_brown_normalized'],
                           y=custom_trendline(common_inat_brown_50['color_value_brown_normalized'], slope=0.902),
                           mode='lines',
                           line=dict(color='green', width=2),
                           name='Actual Correlation (R = 0.902) - AFTER SELECTION')

fig_normalized.update_layout(xaxis_title="Brown Density Normalized",
                            yaxis_title="INaturalist Density Normalized (09-16)",
                            title=f'Brown vs INaturalist Density Normalized, R = {round(correlation_value, 3)}',)
fig_normalized.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.01
))
fig_normalized.show()
```

```python
# difference visualization
max_brown_dv = round(np.max(difference_list)*common_inat_brown_50['color_value_brown'].max(), 1)
mean_brown_dv = round(np.mean(difference_list)*common_inat_brown_50['color_value_brown'].max(), 1)

max_inat_dv = round(np.max(difference_list)*common_inat_brown_50['color_value_inat'].max(), 1)
mean_inat_dv = round(np.mean(difference_list)*common_inat_brown_50['color_value_inat'].max(), 1)

df_diff_visu = pd.DataFrame({'metrics': ['max_visu', 'mean_visu'],
                             'inat': [max_inat_dv, mean_inat_dv], 
                             'brown': [max_brown_dv, mean_brown_dv]})
display(df_diff_visu)

cutoff_pint_diff_inat = 5/(common_inat_brown_50['color_value_inat'].max())
print(f'{cutoff_pint_diff_inat = }')
```

```python
condition = (common_inat_brown_50['differences_pp'] < cutoff_pint_diff_inat)
selected_rows = common_inat_brown_50[condition]
selected_rows.reset_index(inplace=True)
index_selected = list(selected_rows['index'])

# display(selected_rows)

corr_after_selection = selected_rows['color_value_brown_normalized'].corr(selected_rows['color_value_inat_normalized'])
print(f'{corr_after_selection = }')

index_selected = list(selected_rows['index'])
condition_manuelle = []
for i in selection_for_2017on.index:
    condition_manuelle.append((i in index_selected))
condition_finale = pd.Series(condition_manuelle)
selection_for_2017on_final = selection_for_2017on[condition_finale]
# display(selection_for_2017on_final)
selection_for_2017on_final.to_csv('BROWN_finalselection_for_2017on_final.csv', index=None)
```

```python
add_boxes_plot(df_start=selected_rows, title_end=f'Final Selection - Zones in common between INAT-09-16 and Brown&Al, with R = {round(corr_after_selection, 3)}')
```
